```python
# วิธีง่ายๆ - ใช้ IQR (Interquartile Range)
Q1 = df_master['GPA'].quantile(0.25)
Q3 = df_master['GPA'].quantile(0.75)
```

```python
Q1, Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# กรองค่าผิดปกติ
#df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

#กำหนดขอบเขต Lower & Upper

```python
mean_gpa = df_master[(df_master['GPA'] >= lower_bound) & (df_master['GPA'] <= upper_bound)]['GPA'].mean()
```

#หาค่าเฉลี่ยของ gpa

```python
mean_gpa
```

```python
df.loc[(df['GPA'] < lower_bound) | (df['GPA'] > upper_bound), 'GPA'] = round(mean_gpa, 2)
```

#แทนค่า gpa ที่ผิดปกติด้วย mean_gpa

```python
df
```

```python
Q1 = df_master['อายุ'].quantile(0.25)
Q3 = df_master['อายุ'].quantile(0.75)
```

```python
IQR = Q1 + Q3
```

```python
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mean_age = df_master[(df_master['อายุ'] >= lower_bound) & (df_master['อายุ'] <= upper_bound)]['อายุ'].mean()
```

#หา Lower & Upper และ คิดค่าเฉลี่ยของอายุ

```python
mean_age
```

```python
df.loc[(df['อายุ'] < lower_bound) | (df['อายุ'] > upper_bound), 'อายุ'] = round(mean_age, 2)
```
